# Task 1: Load & Explore Data

## Steps
- Load the dataset into a Pandas DataFrame
- Display the first few rows
- Check data types and structure
- Generate summary statistics

## Why This Matters
Exploring the dataset helps identify:
- Column types (numerical, categorical, date)
- Potential data quality issues
- General distribution of data

```python
import pandas as pd

# Load dataset (FIXED encoding issue)
df = pd.read_csv('/content/Superstore.csv', encoding='latin-1')

# 1. Display first few rows
print("First 5 rows:")
display(df.head())

# 2. Check dataset structure
print("\nDataset Info:")
df.info()

# 3. Summary statistics
print("\nSummary Statistics:")
display(df.describe())

# 4. Optional: Include categorical columns too
print("\nFull Summary (including categorical):")
display(df.describe(include='all'))
```

# Task 2: Data Cleaning

## Objective
Prepare the dataset for analysis by handling missing values and correcting data types.

## Steps
- Identify missing values
- Handle missing data
- Convert columns to appropriate data types

## Why This Matters
Clean data ensures:
- Accurate analysis
- Reliable results
- Fewer errors during processing

```python
# Check missing values
df.isnull().sum()

# Handle missing values (example)
df = df.dropna()

# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Verify changes
df.info()
```

# Task 3: Feature Engineering

## Objective
Create new features from existing data to enhance analysis.

## Steps
- Extract Month and Year from Order Date
- Create a new column based on existing numerical columns

## Why This Matters
Feature engineering helps:
- Reveal hidden patterns
- Improve analytical insights
- Make data more useful

```python
# Extract Month and Year
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year

# Create Profit Ratio column
df['Profit Ratio'] = df['Profit'] / df['Sales']

# Verify
df[['Sales', 'Profit', 'Profit Ratio']].head()
```

# Task 4: GroupBy & Aggregation

## Objective
Analyze data by grouping it into meaningful categories.

## Steps
- Calculate total sales by category
- Calculate total profit by region
- Identify top 5 customers based on sales
- Analyze monthly sales trends

## Why This Matters
Grouping allows:
- Identifying high-performing areas
- Understanding customer behavior
- Discovering trends over time

```python
# Sales by category
sales_by_category = df.groupby('Category')['Sales'].sum()
print(sales_by_category)

# Profit by region
profit_by_region = df.groupby('Region')['Profit'].sum()
print(profit_by_region)

# Top 5 customers
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(5)
print(top_customers)

# Monthly sales trend
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum()
print(monthly_sales)
```

# Task 5: Pivot Tables

## Objective
Summarize and analyze data using pivot tables.

## Steps
- Create pivot table for sales by category and region
- Create pivot table for sales trends by segment over time

## Why This Matters
Pivot tables:
- Provide multi-dimensional analysis
- Make comparisons easier
- Help uncover deeper insights

```python
# Pivot table: Sales by Category and Region
pivot1 = pd.pivot_table(df, values='Sales', index='Category', columns='Region', aggfunc='sum')
print(pivot1)

# Pivot table: Sales trends by Segment over time
pivot2 = pd.pivot_table(df, values='Sales', index=['Year', 'Month'], columns='Segment', aggfunc='sum')
print(pivot2)
```
