## Step 2: Data Cleaning

- Fill missing Age values with median
- Fill missing Embarked values with mode
- Drop Cabin column due to many missing values
- Remove duplicate rows

```python
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].median())
titanic["Embarked"] = titanic["Embarked"].fillna(titanic["Embarked"].mode()[0])

titanic = titanic.drop(columns=["Cabin"], errors="ignore")
titanic = titanic.drop_duplicates()
```

### Survival Rate by Gender

This shows the average survival rate for male and female passengers.

```python
titanic.groupby("Sex")["Survived"].mean()
```

### Survival Rate by Class

This shows how passenger class affected survival.

```python
titanic.groupby("Pclass")["Survived"].mean()
```

### Average Age per Class

This calculates the average age of passengers in each class.

```python
titanic.groupby("Pclass")["Age"].mean()
```

### Survival Rate by Age Group

Passengers are grouped into age categories to analyze survival patterns.

```python
titanic["AgeGroup"] = pd.cut(
    titanic["Age"],
    bins=[0, 12, 18, 60, 100],
    labels=["Child", "Teen", "Adult", "Senior"]
)

titanic.groupby("AgeGroup")["Survived"].mean()
```

### Female Passengers Who Survived

Filter all female passengers who survived.

```python
titanic[(titanic["Sex"] == "female") & (titanic["Survived"] == 1)]
```

### Children Who Survived

Filter all children (Age < 12) who survived.

```python
titanic[(titanic["Age"] < 12) & (titanic["Survived"] == 1)]
```

### First Class Passengers Survival Rate

Calculate survival probability of passengers in first class.

```python
titanic[titanic["Pclass"] == 1]["Survived"].mean()
```

## Insights

### Who was more likely to survive?
Females had a significantly higher survival rate than males.

### Did class affect survival?
Yes. First-class passengers had the highest survival rate, while third-class had the lowest.

### Were children prioritized?
Children had a relatively higher survival rate, indicating some prioritization.
